**Predicting if a person would buy life insurnace based on his age using logistic regression**

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
df
```

```python
df.shape##total row and column
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
model.predict_proba(X_test)##not,ha
```

```python
model.score(X_test,y_test)
```

```python
 model.predict(56)
```

```python
# Predict for a single value
value = 56
# Create a DataFrame for the single value with the same feature name
value_df = pd.DataFrame([[value]], columns=['age'])

predicted_class = model.predict(value_df)
predicted_probability = model.predict_proba(value_df)

print(f"Predicted class for age {value}: {predicted_class[0]}")
print(f"Predicted probabilities for age {value}: {predicted_probability[0]}")
```

From the data analysis so far we can conclude that we will use following variables as independant variables in our model
**Satisfaction Level**
**Average Monthly Hours**
**Promotion Last 5 Years**
**Salary**

```python
subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
subdf.head()
```

```python
#Tackle salary dummy variable
```

```python
salary_dummies = pd.get_dummies(subdf.salary, prefix="salary")
salary_dummies
```

```python
df_with_dummies = pd.concat([subdf,salary_dummies],axis='columns')
df_with_dummies.head()
```

```python
df_with_dummies.drop('salary',axis='columns',inplace=True)
df_with_dummies.head()
```

```python
import pandas as pd

# Assuming df_with_dummies is your DataFrame
# Convert boolean columns to integers
bool_columns = df_with_dummies.select_dtypes(include=['bool']).columns
df_with_dummies[bool_columns] = df_with_dummies[bool_columns].astype(int)

# Display the first few rows to check the result
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python
y = df.left
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)  # Increase max_iter as needed
model.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np

# Example confusion matrix

plt.figure(figsize=(5,5))
# Custom function to determine text color based on the value
def annotate_heatmap(data, **kwargs):
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            color = "white" if data[i, j] > np.max(data)/2 else "black"
            plt.text(j+0.5, i+0.5, int(data[i, j]), ha='center', va='center', color=color, **kwargs)

# Plot heatmap without annotations
ax = sn.heatmap(cm, cmap='coolwarm', cbar_kws={'label': 'Scale'}, annot=False, fmt='d')

# Annotate heatmap
annotate_heatmap(cm, fontsize=10)

plt.xlabel('Predicted', fontsize=10)
plt.ylabel('Truth', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
