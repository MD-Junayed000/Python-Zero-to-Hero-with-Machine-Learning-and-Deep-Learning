```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
 digits.target
```

```python
df['target'] = digits.target
df.head()
```

```python
df.shape
```

***Train and the model and prediction***

```python
X = df.drop('target',axis='columns')
y = df.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier ##ensamble term is used to when using multiple algorithm to predict the outcome
model = RandomForestClassifier(n_estimators=20)###default n_estimators=100;use 100 random forest
model.fit(X_train, y_train)
```

```python
model.get_params()
```

```python
# Get detailed information about RandomForestClassifier
help(RandomForestClassifier)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)## y_test hcche ==truth and y_predicted  from model
```

```python
##confusion matrix allow us to plot trth on one axis and prediction on another axis
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)##truth vs predicted
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))##size of x,y
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np

# Example confusion matrix

plt.figure(figsize=(5,5))
# Custom function to determine text color based on the value
def annotate_heatmap(data, **kwargs):
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            color = "white" if data[i, j] > np.max(data)/2 else "black"
            plt.text(j+0.5, i+0.5, int(data[i, j]), ha='center', va='center', color=color, **kwargs)

# Plot heatmap without annotations
ax = sn.heatmap(cm, cmap='coolwarm', cbar_kws={'label': 'Scale'}, annot=False, fmt='d')

# Annotate heatmap
annotate_heatmap(cm, fontsize=10)

plt.xlabel('Predicted', fontsize=10)
plt.ylabel('Truth', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python

from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
iris.target_names
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
X = df.drop('target',axis='columns')
y = df.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier ##ensamble term is used to when using multiple algorithm to predict the outcome
model = RandomForestClassifier(n_estimators=50)###default n_estimators=100;use 100 random forest
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model = RandomForestClassifier(n_estimators=40)
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
