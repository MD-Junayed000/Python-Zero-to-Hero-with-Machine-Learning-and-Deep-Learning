```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\ASUS\Downloads\archive (1)\EEG_data.csv")
data = pd.read_csv(r"C:\Users\ASUS\Downloads\archive (1)\demographic_info.csv")
df
```

```python
# Renaming columns in the 'data' DataFrame
data = data.rename(columns={'subject ID': 'SubjectID', ' gender': 'gender', ' age': 'age', ' ethnicity': 'ethnicity'})

# Merging 'df' with the 'data' DataFrame based on 'SubjectID'
df = df.merge(data, how='inner', on='SubjectID')

# Displaying the first few rows of the merged DataFrame
df.head()
```

```python
#1. Plot the Raw EEG Signal for Each Student (Time vs. Signal)
# Assuming 'SubjectID' and 'VideoID' are in the df DataFrame
subject = 0  # Choose subject 0
video = 0  # Choose video 0

# Filter data for the chosen subject and video
subject_video_data = df[(df['SubjectID'] == subject) & (df['VideoID'] == video)]

# Plot raw EEG signal (example with Delta band)
plt.figure(figsize=(10, 6))
plt.plot(subject_video_data['Delta'], label='Delta Band Power')
plt.xlabel('Time (s)')
plt.ylabel('Delta Power')
plt.title(f'Raw EEG Signal for Subject {subject} while watching Video {video}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2. Plot Power Spectral Density (PSD) for Each Frequency Band
from scipy import signal

# Assuming Delta data as a time series (you can select any frequency band)
eeg_data = subject_video_data['Delta'].values

# Plotting PSD using Welch's method
frequencies, psd = signal.welch(eeg_data, fs=1.0, nperseg=256)

plt.figure(figsize=(10, 6))
plt.semilogy(frequencies, psd)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density (ÂµV^2/Hz)')
plt.title(f'Power Spectral Density for Subject {subject} while watching Video {video}')
plt.grid(True)
plt.show()
```

```python
#3. Time-Frequency Representation (Spectrogram)
# Plotting Spectrogram
plt.figure(figsize=(10, 6))
plt.specgram(eeg_data, Fs=1.0, NFFT=256, noverlap=128, cmap='plasma')
plt.title(f'Spectrogram for Subject {subject} while watching Video {video}')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Power (ÂµV^2)')
plt.show()
```

```python
#4. Frequency Band Power (Delta, Theta, Alpha, Beta, Gamma)
plt.figure(figsize=(12, 8))

# Plotting power of each frequency band
bands = ['Delta', 'Theta', 'Alpha1', 'Alpha2', 'Beta1', 'Beta2', 'Gamma1', 'Gamma2']
for band in bands:
    plt.plot(subject_video_data[band], label=band)

plt.xlabel('Time (s)')
plt.ylabel('Power')
plt.title(f'Frequency Band Power for Subject {subject} while watching Video {video}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5. EEG Power Comparison Between Confusing and Non-Confusing Videos
# Create two categories based on confusion label (assuming 'predefinedlabel' is the confusion label)
confusing = df[df['predefinedlabel'] == 7]  # Most confusing videos
non_confusing = df[df['predefinedlabel'] == 1]  # Least confusing videos

# Plotting comparison of EEG power for confusing vs. non-confusing videos
plt.figure(figsize=(12, 8))
plt.plot(confusing['Delta'], label='Confusing - Delta')
plt.plot(non_confusing['Delta'], label='Non-Confusing - Delta')

plt.xlabel('Time (s)')
plt.ylabel('Power')
plt.title('EEG Power Comparison for Confusing vs Non-Confusing Videos (Delta Band)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6. Correlation Between EEG Power and Confusion Rating
# Calculate correlation between Delta power and confusion rating
corr = df['Delta'].corr(df['predefinedlabel'])

plt.figure(figsize=(10, 6))
plt.scatter(df['predefinedlabel'], df['Delta'], label='Delta Power')
plt.xlabel('Confusion Rating')
plt.ylabel('Delta Power')
plt.title(f'Correlation Between Delta Power and Confusion Rating: {corr:.2f}')
plt.grid(True)
plt.show()
```

```python
#7. Clustering EEG Data Based on Frequency Bands (Optional)
from sklearn.cluster import KMeans

# Use frequency bands data for clustering
frequency_band_data = df[['Delta', 'Theta', 'Alpha1', 'Alpha2', 'Beta1', 'Beta2', 'Gamma1', 'Gamma2']]

# Apply k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(frequency_band_data)

# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Delta'], df['Theta'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Delta Power')
plt.ylabel('Theta Power')
plt.title('Clustering of EEG Data Based on Frequency Bands')
plt.colorbar(label='Cluster')
plt.show()
```

```python
#10. Relationship Between Attention Levels and Confusion Labels
plt.figure(figsize=(10, 6))
plt.scatter(df['Attention'], df['predefinedlabel'], color='purple')
plt.xlabel('Attention Level')
plt.ylabel('Confusion Rating')
plt.title('Attention vs Confusion Rating')
plt.grid(True)
plt.show()
```
