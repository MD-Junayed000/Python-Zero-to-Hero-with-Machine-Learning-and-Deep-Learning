#  Merge Dataframes

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.concat([df1,df2],axis=1)
df3
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

# Type Of DataBase Joins

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")##takes intersection only
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3###takes union
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

# pd.get_dummies()~

*** Basic Examples
 Example 1: One-hot encode a single column***

```python
import pandas as pd

df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green']})
print("Original DataFrame:")
print(df)

dummies = pd.get_dummies(df)
print("\nAfter get_dummies:")
print(dummies)
```

*** Example 2: Multiple columns***

```python
df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green'],
    'Size': ['S', 'M', 'L']
})
dummies = pd.get_dummies(df)
print(dummies)
```

```python
demo=pd.get_dummies(df, drop_first=True)
demo
```

```python
trail=pd.get_dummies(df, columns=['Color'])
trail
```
