# Canny Edge Detection

```python
#import the dependencies
import cv2
import numpy as np


image = cv2.imread(r"C:\Users\ASUS\Downloads\valid_real\real\valid_real_0015401.png")
image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# performing the edge detetcion
gradients_sobelx = cv2.Sobel(image, -1, 1, 0)
gradients_sobely = cv2.Sobel(image, -1, 0, 1)
gradients_sobelxy = cv2.addWeighted(gradients_sobelx, 0.5, gradients_sobely, 0.5, 0)

gradients_laplacian = cv2.Laplacian(image, -1)

canny_output = cv2.Canny(image, 80, 150)

cv2.imshow('Sobel x', gradients_sobelx)
cv2.imshow('Sobel y', gradients_sobely)
cv2.imshow('Sobel X+y', gradients_sobelxy)
cv2.imshow('laplacian', gradients_laplacian)
cv2.imshow('Canny', canny_output)
cv2.waitKey()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread(r"C:\Users\ASUS\Downloads\valid_real\real\valid_real_0015401.png", cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(img, cv2.CV_64F, ksize=3)
lap = np.uint8(np.absolute(lap))
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
edges = cv2.Canny(img,100,200)

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX, sobelY)

titles = ['image', 'Laplacian', 'sobelX', 'sobelY', 'sobelCombined', 'Canny']
images = [img, lap, sobelX, sobelY, sobelCombined, edges]
for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

#  Denoising and edge detection using opencv in Python

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Corrected line below
img = cv2.imread(r"C:\Users\ASUS\Downloads\valid_real\real\valid_real_0015401.png")  

# Kernel for convolution
kernel = np.ones((5, 5), np.float32) / 25  # Kernel is a small matrix

# Different filtering techniques
filt_2D = cv2.filter2D(img, -1, kernel)    # Convolution using the kernel we provide



blur = cv2.blur(img, (5, 5))              # Convolution with a normalized filter
blur_gaussian = cv2.GaussianBlur(img, (5, 5), 0)  # Gaussian kernel is used
median_blur = cv2.medianBlur(img, 5)      # Using kernel size 5, better on edges
bilateral_blur = cv2.bilateralFilter(img, 9, 75, 75)  # Good for noise removal

# Display the images
cv2.imshow("Original", img)
cv2.imshow("2D Custom Filtered", filt_2D)
cv2.imshow("Blur", blur)
cv2.imshow("Gaussian Blur", blur_gaussian)
cv2.imshow("Median Blur", median_blur)
cv2.imshow("Bilateral", bilateral_blur)
cv2.waitKey(0)          
cv2.destroyAllWindows()
######################
```
